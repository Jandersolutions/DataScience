```python
# retornando a mensalidade mais cara
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
# retornando a mensalidade mais baratas
df[df['mensalidade'] == df['mensalidade'].min()]
```

```python
# vendo a oferta de cursos presencias e a distancia
# A grande maioria dos cursos ofertados em 2018 foram cursos a distancia, seguido dos cursos no turno noturno
df['turno'].value_counts()
```

```python
# exibindo a cidade e unidade federativa com as mensalidades mais baixas
df[df['mensalidade'] == df['mensalidade'].min()][['uf_busca','cidade_busca']]
```

```python
# exibindo a mensalidade mais barato por cidade
df[df['mensalidade'] == df['mensalidade'].min()]['cidade_busca']
```

```python
# mensalidade mais cara por unidade federativa
df.groupby('uf_busca')['mensalidade'].max()
```

```python
# mensalidade mais barata por unidade federativa
df.groupby('uf_busca')['mensalidade'].min()
```
