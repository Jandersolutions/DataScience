<h3><i class="fas fa-chevron-circle-right"></i>1- Calcule a pontuação softmax de ‘sepal length’:</h3>

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
sepallength = np.genfromtxt(url, delimiter=",", dtype="float", usecols=[0])
sepallength
```

```python
def softmax(sepallength):
    e_sepallength = np.exp(sepallength - np.max(sepallength))
    return e_sepallength / e_sepallength.sum()
scores = [3.0, 1.0, 0.2]
print(softmax(scores))
```
